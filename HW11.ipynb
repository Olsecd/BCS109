```
# load pandas and numpy libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/Olsecd/BCS109/master/Day1_Data%20-%20M2_Data%20(1).csv.csv')
df.head()
```

```
#check for nulls
df.isnull().sum()
```

```
normalized_df=(df-df.mean())/df.std()
for col in df:
    sns.kdeplot(normalized_df[col], shade=True)
```

```
corr = df.corr()
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(data=corr, square=True , annot=True, cbar=True, ax = ax)
```
